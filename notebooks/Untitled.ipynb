```python
import pandas as pd
from sqlalchemy import create_engine, text
```

```python
database_name = 'prescribers1'   

connection_string = f"postgresql://postgres:postgres@localhost:5432/{database_name}"
```

```python
engine = create_engine(connection_string)
```

```python
query1 = """

SELECT
    *
FROM
    zip_fips
    
"""

with engine.connect() as connection:
    result = connection.execute(text(query1))
    zip_fips = pd.DataFrame(result.fetchall(), columns=result.keys())
zip_fips.head   
```

```python
query2 = """

SELECT
    *
FROM
    fips_county
    
"""

with engine.connect() as connection:
    result = connection.execute(text(query2))
    fips_county = pd.DataFrame(result.fetchall(), columns=result.keys())
fips_county.head 
```

```python
query3 = """

SELECT
    *
FROM
    drug
    
"""

with engine.connect() as connection:
    result = connection.execute(text(query3))
    drug = pd.DataFrame(result.fetchall(), columns=result.keys())
drug.head 
```

```python
query4 = """

SELECT
    *
FROM
    prescription
    
"""

with engine.connect() as connection:
    result = connection.execute(text(query4))
    prescription = pd.DataFrame(result.fetchall(), columns=result.keys())
prescription.head 
```

```python
query5 = """

SELECT
    *
FROM
    overdose_deaths
    
"""

with engine.connect() as connection:
    result = connection.execute(text(query5))
    overdose_deaths = pd.DataFrame(result.fetchall(), columns=result.keys())
overdose_deaths.head  
```

```python
query6 = """

SELECT
    *
FROM
    prescriber
    
"""

with engine.connect() as connection:
    result = connection.execute(text(query6))
    prescriber = pd.DataFrame(result.fetchall(), columns=result.keys())
prescriber.head 
```

```python
#Identify Opioids
```

```python
# Filter opioids
opioid_drugs = drug[drug['opioid_drug_flag'] == 'Y'][['drug_name', 'long_acting_opioid_drug_flag']]
```

```python
#Filter to TN
```

```python
presc_with_county = (
    prescription
    .merge(
        prescriber[['npi', 'nppes_provider_state', 'nppes_provider_zip5']],
        on='npi',
        how='left'
    )
)
presc_with_county = presc_with_county[presc_with_county['nppes_provider_state'] == 'TN']
```

```python
#Join zipcode and county
```

```python
presc_with_county = presc_with_county.merge(
    zip_fips[['zip', 'fipscounty']],
    left_on='nppes_provider_zip5',
    right_on='zip',
    how='left')
```

```python
#Group prescriptions by drug and county
```

```python
county_drug_presc = (
    presc_with_county.groupby(['drug_name', 'fipscounty'])
    .agg({'total_claim_count': 'sum'})
    .reset_index()
)
```

```python
#Deaths by county
```

```python
county_overdose = overdose_deaths.groupby('fipscounty').agg({'overdose_deaths': 'sum'}).reset_index()
```

```python
#Merge presc with deaths
```

```python
county_drug_presc['fipscounty'] = county_drug_presc['fipscounty'].astype(str).str.zfill(5)
county_overdose['fipscounty'] = county_overdose['fipscounty'].astype(str).str.zfill(5)
```

```python
analysis = county_drug_presc.merge(county_overdose, on='fipscounty', how='inner')
```

```python
analysis['total_claim_count'] = analysis['total_claim_count'].astype(float)
analysis['overdose_deaths'] = analysis['overdose_deaths'].astype(float)
```

```python
county_drug_presc.columns
```

```python
opioid_counts = county_drug_presc.groupby('drug_name').agg(
    total_prescriptions=('total_claim_count', 'sum')
).reset_index()
```

```python
total_deaths = county_overdose['overdose_deaths'].sum()
```

```python
#Add total deaths
```

```python
opioid_summary = opioid_counts.copy()
opioid_summary['total_overdose_deaths'] = total_deaths
```

```python
#Sort by total prescriptions
```

```python
opioid_summary = opioid_summary.sort_values(by='total_prescriptions', ascending=False)
```
